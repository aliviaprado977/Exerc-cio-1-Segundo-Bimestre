```python
import requests
from bs4 import BeautifulSoup
import sqlite3
```

```python
url = 'https://www.passandoaregua.com.br/'

req = requests.get(url)
html = req.text
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
noticias = []

for item in soup.find_all('a'):
    titulo = item.text.strip()
    link = item.get('href')

    if titulo != '' and link != None and '/noticias/' in link and len(titulo) > 20:
        noticias.append({
            'titulo': titulo,
            'link': link
        })
```

```python
conn = sqlite3.connect('noticias.db')
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS noticias(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    titulo TEXT NOT NULL,
    link TEXT NOT NULL
)
""")
```

```python
import sqlite3

conn = sqlite3.connect('noticias.db')
cursor = conn.cursor()

cursor.execute("DELETE FROM noticias")

for n in noticias:
    cursor.execute(
        "INSERT INTO noticias (titulo, link) VALUES (?, ?)",
        (n['titulo'], n['link'])
    )

conn.commit()

cursor.execute("SELECT * FROM noticias LIMIT 10")
dados = cursor.fetchall()

for d in dados:
    print(d)

conn.close()
```
